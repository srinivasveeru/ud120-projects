```python
def areaShape(shape):
  print("Received shape is: ",shape)
  match shape:
    case "square":
      side=int(input("Enter the side of the square:"))
      area = side * side
    case "rectangle":
      length=int(input("Enter the length of the rectangle:"))
      width=int(input("Enter the width of the rectangle:"))
      area = length * width
    case "circle":
      radius=int(input("Enter the radius of the circle:"))
      area = 3.14 * radius * radius
    case "triangle":
      base=int(input("Enter the base of the triangle:"))
      height=int(input("Enter the height of the triangle:"))
      area = 0.5 * base * height
    case _:
      print("Invalid shape")
      area = None # Assign None or raise an error for invalid shapes
  return area

print("Enter the shape (square, rectangle, circle, triangle): ")
shape = input()
area = areaShape(shape)
if area is not None:
  print("The area of the", shape, "is", area)
else:
  print("Could not calculate area for invalid shape.")
```
