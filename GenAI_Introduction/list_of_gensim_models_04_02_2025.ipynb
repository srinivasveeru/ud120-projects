```python
f=open("Test.txt","r")
lines=0
words=0
chars=0

for line in f:
  lines += 1
  words += len(line.split())
  chars += len(line)

print("Number of lines: ",lines)
print("Number of words: ",words)
print("Number of characters: ",chars)
f.close()

```
