```python
def karrmax(arr,k):
  #subarr=[]
  karr=[]
  for i in range(len(arr)-(k-1)):
    #j = i
    #print(i)
    subarr = []
    for j in range(i,i+k):
      subarr.append(arr[j])
      #print(subarr)
      #print(j)
    karr.append(max(subarr))
  print(karr)


arr=[1, 2, 3, 1, 4, 5, 2, 3, 6, 1, 2, 3, 1, 4, 5, 2, 3, 6]
karrmax(arr,6)

```

```python
def DecToBin(n):
  bin = ''
  rbin=''
  i=n
  if(n>0):
    while(i>0):
      bin+=str(i%2)
      i=int(i/2)
      #if(i==1):
       # break
      print(i)
  else:
    bin = 0
  m=len(bin)
  for k in range(m):
    rbin+=bin[m-k-1]
  return rbin

DecToBin(20)
```
